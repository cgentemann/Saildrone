```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np

adir_data = 'F:/data/cruise_data/saildrone/sss/sss_collocations/'
saildrone_filenames = ['saildrone-gen_5-atomic_eurec4a_2020-sd1026-20200117T000000-20200302T235959-1_minutes-v1.1589306725934rssv04.0_filesave4.nc',
        'saildrone-gen_5-atomic_eurec4a_2020-sd1026-20200117T000000-20200302T235959-1_minutes-v1.1589306725934jplv04.3_filesave4.nc',
        'saildrone-gen_5-atomic_eurec4a_2020-sd1060-20200117T000000-20200302T235959-1_minutes-v1.1589306886594rssv04.0_filesave4.nc']

plt.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
ds=[]
for iusv in range(3):
    fname=adir_data+saildrone_filenames[iusv]
    ds_usv=xr.open_dataset(fname)#.isel(trajectory=0).swap_dims({'obs':'time'})
    ds_usv.close()
    xlat=ds_usv.lat
    xlon=ds_usv.lon
    dkm2 = abs(np.abs((((xlon[1:].data-xlon[0:-1].data)**2+(xlat[1:].data-xlat[0:-1].data)**2)**.5)*110.567*np.cos(np.pi*xlat[1:].data/180)))
    dkm2=np.append(dkm2,dkm2[66238]) #add on last point
    dkm3 = dkm2.cumsum()
    ds_usv['dist_total']=xr.DataArray(dkm3,dims=('time'),coords={'time':ds_usv.time})
    ds_usv['dist_between']=xr.DataArray(dkm2,dims=('time'),coords={'time':ds_usv.time})
    if iusv==1:
        ds_usv['smap_SSS_40km']=ds_usv.smap_SSS*np.nan
    ds_usv=ds_usv.where(ds_usv.smap_SSS>10)
    ds.append(ds_usv)
ds_saildrone = xr.concat(ds,dim='trajectory')
```
